```python
# Load the Retail data

retail = pd.read_csv(r"E:\AIML INTERNSHIP\Datasets\Walmart_01.csv", header = 0)

# Copy to back-up files

retail_bk = retail.copy()

# Display first five claims data

retail.head()
```

```python
# Create new variabl as 'ws_range'

retail['ws_range'] = pd.cut(retail['Weekly_Sales'], bins=5, labels=['very_low', 'low', 'medium', 'high', 'very_high'])

# Display the sample 5 records

retail.sample(10)
```

```python
# Create new variabl as 'ws_range_01'

retail['ws_range_01'] = pd.cut(retail['Weekly_Sales'], bins=3, labels=['low', 'medium', 'high'])

# Display the sample 5 records

retail.sample(10)
```

```python
# Display the 'ws_range_01' value by count

retail['ws_range_01'].value_counts()
```

```python
# Display the 'ws_range' value by count

retail['ws_range'].value_counts()
```

```python
# Convert the object format 'Date' to DATE format

retail['Date'] = pd.to_datetime(retail['Date'])

# display first five records

retail.head()
```

```python
# Year from the Date

retail['year1'] = pd.DatetimeIndex(retail['Date']).year

# Month from the Date

retail['month1'] = pd.DatetimeIndex(retail['Date']).month

# Day from the Date

retail['Day1'] = pd.DatetimeIndex(retail['Date']).day

# Display first five records

retail.head()
```
